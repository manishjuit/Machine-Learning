# Feature Engineering

```python
attrition_data
```

```python
attrition_data['Department'].value_counts()
```

lets convert categorical variable into numerical

```python
attrition_numerical_data=pd.get_dummies(attrition_data,columns=['Attrition','Department','EducationField','MaritalStatus'],drop_first=True)
```

```python
attrition_numerical_data.columns
```

```python
X_feature_list=['Age', 'DistanceFromHome', 'Education', 'EnvironmentSatisfaction',
       'JobSatisfaction', 'MonthlyIncome', 'NumCompaniesWorked',
       'WorkLifeBalance', 'YearsAtCompany',
       'Department_Research & Development', 'Department_Sales',
       'EducationField_Life Sciences', 'EducationField_Marketing',
       'EducationField_Medical', 'EducationField_Other',
       'EducationField_Technical Degree', 'MaritalStatus_Married',
       'MaritalStatus_Single']
```

```python
X_feature=attrition_numerical_data[X_feature_list]
```

```python
Y_feature=attrition_numerical_data[['Attrition_Yes']]
```

Let's include all the feature first check for the accuracy

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X_feature,Y_feature,random_state=101)
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic=LogisticRegression()
logistic.fit(X_train,Y_train)
```

```python
logistic.coef_
```

```python
logistic.intercept_
```

```python
Y_prediction=logistic.predict(X_test)
```

```python
score = logistic.score(X_test, Y_test)
print(score)
```

```python
from sklearn import metrics
```

```python
cm = metrics.confusion_matrix(Y_test, Y_prediction)
print(cm)
```

```python
Y_test['preds'] =Y_prediction
```
